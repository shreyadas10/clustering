 1. What is unsupervised learning in the context of machine learning ?

 -> In the context of machine learning, unsupervised learning is a type of algorithm that analyzes and clusters unlabeled data to discover patterns, relationships, and groupings without explicit guidance. Unlike supervised learning, which relies on labeled data to make predictions, unsupervised learning aims to understand the underlying structure and distribution of data.

2. How does K-Means clustering algorithm work ?

-> K-means clustering is a machine learning algorithm that groups data points into clusters based on similarity. It works by iteratively assigning data points to the nearest cluster centroid and then recalculating the centroids based on the new assignments. The goal is to minimize the distance between data points and their assigned cluster centers.

3. Explain the concept of a dendrogram in hierarchical clustering ?

-> A dendrogram is a tree-like diagram that visually represents the hierarchy of clusters generated by hierarchical clustering algorithms. It shows how data points or clusters are grouped together based on their similarity or distance, with branches indicating the hierarchy. The height of the branches often reflects the distance between the merged clusters.

4. What is the main difference between K-Means and Hierarchical Clustering ?

-> The main difference between K-means and hierarchical clustering lies in how they create clusters. K-means is a centroid-based algorithm that requires you to specify the number of clusters in advance, while hierarchical clustering builds a hierarchy of clusters, allowing you to determine the optimal number of clusters later.


5. What are the advantages of DBSCAN over K-Means?

-> DBSCAN offers several advantages over K-Means, particularly when dealing with complex data structures. DBSCAN is better at identifying clusters of arbitrary shapes, handling noise and outliers effectively, and doesn't require the user to predefine the number of clusters. K-Means, on the other hand, is simpler to implement, can handle large datasets, and is well-suited for spherical or compact clusters.

6. When would you use Silhouette Score in clustering ?

-> The Silhouette Score is used in clustering to evaluate the quality of the clusters formed by a clustering algorithm, particularly for algorithms like K-Means. It helps determine if data points are well-clustered, meaning they are tightly grouped within their own cluster and distinct from other clusters. Essentially, it measures how well each data point fits within its assigned cluster compared to other clusters

7. What are the limitations of Hierarchical Clustering ?

-> Hierarchical clustering, while powerful, has limitations including high computational complexity, sensitivity to outliers and noise, and difficulties in handling large datasets and mixed data types. The irreversible merging of clusters and the need for arbitrary decisions about stopping points also pose challenges.

8.  Why is feature scaling important in clustering algorithms like K-Means ?

->  Feature scaling is crucial in clustering algorithms like K-Means because it ensures that all features contribute equally to the distance calculations, preventing features with larger ranges from dominating the clustering process.

9. How does DBSCAN identify noise points ?

->The algorithm works by defining clusters as dense regions separated by regions of lower density. This approach allows DBSCAN to discover clusters of arbitrary shape and identify outliers as noise

10.  Define inertia in the context of K-Means ?

-> It is calculated by measuring the distance between each data point and its centroid, squaring this distance, and summing these squares across one cluster. A good model is one with low inertia AND a low number of clusters ( K ). However, this is a tradeoff because as K increases, inertia decreases.

11. What is the elbow method in K-Means clustering ?

->  The Elbow Method is a heuristic used in k-means clustering to determine the optimal number of clusters (k). It involves plotting the Within-Cluster Sum of Squares (WCSS) for various k values and visually identifying an "elbow" point on the graph. The "elbow" represents the point where the rate of decrease in WCSS significantly slows down, indicating that adding more clusters doesn't substantially improve the clustering.

12. Describe the concept of "density" in DBSCAN

-> In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), "density" refers to the number of data points within a given radius around a specific point. A cluster is defined as a dense region of points separated by regions of lower density.

-> Clustering Categorical(or mixed) Data in R | by Maryam ...Yes, hierarchical clustering can be used for categorical data. However, it requires using appropriate distance or dissimilarity metrics designed for categorical data rather than the standard Euclidean distance used for numerical data.

14. What does a negative Silhouette Score indicate ?

15. Explain the term "linkage criteria" in hierarchical clustering

-> Single-Link Hierarchical Clustering Clearly Explained ...In hierarchical clustering, "linkage criteria" or "linkage methods" determine how the distance between two clusters is calculated. They dictate how the algorithm merges clusters based on their similarity. Essentially, linkage criteria define the definition of "distance" between clusters, impacting the final clustering structure.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities ?

-> K-Means performs poorly on data with varying cluster sizes or densities because it assumes clusters are spherical and have similar sizes and densities.

17. What are the core parameters in DBSCAN, and how do they influence clustering ?

->  DBSCAN, a density-based clustering algorithm, relies on two core parameters: Epsilon (Îµ) and MinPts. Epsilon defines the radius around a data point within which other points are considered its neighbors, while MinPts specifies the minimum number of points within the epsilon radius required to form a dense region (a cluster). These parameters directly influence cluster formation, impacting both the number and shape of the clusters identified.

18. How does K-Means++ improve upon standard K-Means initialization

->  Spreading out the initial centroids more evenly across the data, reducing the likelihood of poor initialization and improving the chances of converging to a better clustering solution.

It does this by using a probabilistic approach to select initial centroids that are farther apart from each other.

19. What is agglomerative clustering ?

-> Agglomerative clustering is a hierarchical clustering technique that uses a "bottom-up" approach to group data points into clusters. It starts with each data point being its own individual cluster and then iteratively merges the most similar clusters until a single, large cluster containing all data points is formed. This process creates a hierarchy of clusters, which can be visualized using a dendrogram.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

-> The Silhouette Score is a more comprehensive and reliable metric for evaluating clustering models than inertia alone. Inertia, also known as distortion, only considers the compactness within clusters, while Silhouette Score assesses both compactness and separation between clusters. This nuanced approach allows for a more balanced assessment of clustering quality, particularly when dealing with datasets where clusters may have varying sizes or shapes.

21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

x,y = make_blobs(n_samples= 1000, n_features= 2, centers= 4 , random_state= 1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.30, random_state= 1)
kmeans = KMeans(n_clusters= 4, random_state= 1)
kmeans.fit(x_train)
y_labels  = kmeans.predict(x_train)
plt.figure(figsize=(10,8))
plt.scatter(x_train[:,0],x_train[0:,1], c = y_labels)
plt.title('K-Means Clustering')
plt.show()
```

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

```python

from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=42)

k_values = range(2, 6)
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='skyblue', edgecolor='black')
plt.title("Silhouette Scores for KMeans Clustering")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.xticks(k_values)
plt.ylim(0, 1)
plt.grid(True, axis='y')
plt.show()
```

31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries

```python

X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
grid = np.c_[xx.ravel(), yy.ravel()]
Z = kmeans.predict(grid)
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, cmap='Pastel1', alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Set1', edgecolor='k', s=40)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='black', label='Centroids')
plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

```python

from sklearn.metrics import silhouette_samples, silhouette_score

iris = load_iris()
X = iris.data
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

sample_silhouette_values = silhouette_samples(X, labels)

plt.figure(figsize=(10, 6))
y_lower = 10
for i in range(3):  # For each cluster
    ith_cluster_silhouette_values = sample_silhouette_values[labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      alpha=0.7, label=f"Cluster {i}")

    y_lower = y_upper + 10  # 10 for spacing between clusters

avg_score = silhouette_score(X, labels)
plt.axvline(avg_score, color="red", linestyle="--", label=f"Avg: {avg_score:.2f}")

plt.title("Silhouette Coefficients per Sample (Iris Dataset)")
plt.xlabel("Silhouette Coefficient Value")
plt.ylabel("Sample Index (Clustered)")
plt.legend()
plt.grid(True)
plt.show()
```

46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
features)

```python

import seaborn as sns

wine = load_wine()
X = wine.data[:, :4]  # First 4 features
feature_names = wine.feature_names[:4]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

df = pd.DataFrame(X, columns=feature_names)
df['Cluster'] = labels.astype(str)  # Convert to string for better color mapping

sns.pairplot(df, hue='Cluster', diag_kind='kde', palette='Set1')
plt.suptitle("KMeans Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count

```python

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

noise_points = np.random.uniform(low=-10, high=10, size=(30, 2))
X = np.vstack([X, noise_points])

dbscan = DBSCAN(eps=1.5, min_samples=5)
labels = dbscan.fit_predict(X)

n_noise = np.sum(labels == -1)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {n_noise}")
```
